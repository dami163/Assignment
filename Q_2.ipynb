```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('highlights_and_shadows.jpg', cv.IMREAD_COLOR)

gamma = .6

y = np.array(
    [(i/255.0)**(gamma)*255.0 for i in np.arange(0, 256)]).astype('uint8')

img_g = cv.LUT(img, y)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_g = cv.cvtColor(img_g, cv.COLOR_BGR2RGB)

f, axarr = plt.subplots(1, 2)
axarr[0].imshow(img)
axarr[0].set_title('Original')
axarr[1].imshow(img_g)
axarr[1].set_title('Gamma Corrected')
```

```python


fig, ax = plt.subplots()
ax.plot(y)
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_aspect('equal')

plt.suptitle("Gamma correction Value Curve")
plt.show()
```

```python


f, ax = plt.subplots(1, 2)
color = ('b', 'g', 'r')

for i, c in enumerate(color):
    hist_orig = cv.calcHist([img], [i], None, [256], [0, 256])
    ax[0].plot(hist_orig, color=c)
    ax[0].set_title('Original Histrogram')
    hist_gamma = cv.calcHist([img_g], [i], None, [256], [0, 256])
    ax[1].plot(hist_gamma, color=c)
    ax[1].set_title('Gamma Corrected Histrogram')
```
